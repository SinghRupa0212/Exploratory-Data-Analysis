```python
from sklearn.preprocessing import RobustScaler
```

```python
Features=Data.drop(Data[['target']], axis=1)
Features.head()
```

```python
Target=Data.target
Target.head()
```

```python
RS = RobustScaler()
```

```python
X_Data= RS.fit_transform(Features)
X_Data
```

### Without Train, Test and Split

```python
from sklearn.linear_model import LogisticRegression
```

```python
Lr = LogisticRegression()
```

```python
Lr.fit(X_Data, Target)
```

```python
Lr.score(X_Data, Target)
```

### With Train and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(X_Data,Target,test_size=0.18)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
Model= LogisticRegression(solver='lbfgs')
```

```python
Model.fit(X_train,Y_train)
```

```python
Train_Score=Model.score(X_train,Y_train)
Train_Score
```

```python
Test_Score=Model.score(X_train,Y_train)
Test_Score
```

```python
New_Pred=Model.predict(X_test)
New_Pred
```

```python
pd.DataFrame({'Actual_Data': Y_test, 'New_Prediction': New_Pred}).head()
```

```python
Model.predict_proba(X_test)
```

### Model Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
Cm= confusion_matrix(Y_test, New_Pred)
```

```python
Cm
```

```python
sns.heatmap(Cm,annot=True)
```
